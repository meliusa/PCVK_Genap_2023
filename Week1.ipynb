Gunakan beberapa library berikut sebagai langkah pertama:

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

Langkah 2 untuk membaca dan menampilkan image 


```python
# Membuat list untuk menyimpan url dari bberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/777"]

# Baca dan tampilkan image
# Loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          # read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) # resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_RGB2BGR)  # convert color to RGB
  final_frame = cv.hconcat((image, image_2))      # concatenate image
  cv2_imshow(final_frame)                         # show image
  print('\n')
```

Langkah 3 pada praktikum ini adalah melihat ukuran file image, dengan cara sebagai
berikut:

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

Langkah 4 berikut digunakan untuk mengakses pixel dengan memberikan garis
horizontal berwarna putih di tengah image 

```python
image_2 = cv.cvtColor(image, cv.COLOR_RGB2BGR)
image_3 = cv.cvtColor(image, cv.COLOR_RGB2BGR)

# Membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

3. Baris kode image = cv.resize(image, (0,0), fx=0.5, fy=0.5) merupakan sebuah perintah pada OpenCV untuk merubah ukuran gambar (resize image). Berikut penjelasan dari setiap argumen pada fungsi cv.resize():

* image: variabel yang berisi gambar yang akan diresize.

* (0, 0): ukuran output gambar yang baru. Kita bisa menentukan ukuran output gambar secara manual dengan mengganti nilai (0, 0) dengan ukuran pixel yang diinginkan, seperti (800, 600).

* fx=0.5, fy=0.5: faktor skala yang digunakan untuk merubah ukuran gambar. fx dan fy adalah faktor skala pada sumbu horizontal dan vertikal. Dalam contoh ini, gambar akan diresize dengan faktor skala sebesar 0.5 pada kedua sumbu, sehingga ukuran gambar akan menjadi separuh dari ukuran gambar aslinya.

> Dalam kasus ini, gambar yang diresize akan disimpan dalam variabel image yang sama, sehingga gambar asli akan diganti dengan gambar yang telah diresize. Hal ini berguna untuk menghemat memori dan menghindari duplikasi variabel gambar.

4. [255, 255, 255] pada baris kode image_3[int((tinggi)/2),y] = [255,255,255] adalah sebuah nilai warna dalam format BGR (blue-green-red). Nilai [255, 255, 255] ini merepresentasikan warna putih pada gambar.

> Baris kode tersebut mengubah nilai pixel pada koordinat (int((tinggi)/2), y) pada gambar image_3 menjadi warna putih. int((tinggi)/2) mengembalikan nilai integer dari tinggi yang dibagi 2, sehingga baris kode tersebut mengubah nilai pixel pada baris tengah gambar.

> Tujuan dari baris kode tersebut bisa bermacam-macam tergantung dari konteks program yang lebih besar. Sebagai contoh, jika gambar yang sedang diproses adalah sebuah citra medis, maka baris kode tersebut mungkin digunakan untuk menandai tengah-tengah gambar sebagai suatu referensi dalam proses analisis citra medis.

5. Pixel dan resolusi gambar memiliki keterkaitan erat. Pixel adalah unit terkecil yang membentuk gambar digital, sedangkan resolusi gambar adalah jumlah total pixel pada gambar.

> Resolusi gambar yang tinggi berarti gambar memiliki jumlah pixel yang banyak, sedangkan resolusi gambar yang rendah berarti gambar memiliki jumlah pixel yang sedikit. Semakin tinggi resolusi gambar, semakin banyak pixel yang ada pada gambar, dan semakin detail dan tajam gambar tersebut. Sebaliknya, semakin rendah resolusi gambar, semakin sedikit pixel yang ada pada gambar, dan semakin kabur atau pecah gambar tersebut.

> Keterkaitan antara pixel dan resolusi gambar dapat dihitung dengan menggunakan rumus berikut: Resolusi gambar = Jumlah pixel pada tinggi x jumlah pixel pada lebar. Sebagai contoh, jika sebuah gambar memiliki ukuran 800 x 600 pixel, maka resolusi gambar adalah 480.000 pixel (800 x 600).

> Oleh karena itu, ketika kita ingin memiliki gambar yang tajam dan detail, maka kita perlu menggunakan gambar dengan resolusi yang tinggi dan jumlah pixel yang banyak. Namun, semakin banyak jumlah pixel pada gambar, maka semakin besar ukuran file gambar dan semakin membutuhkan ruang penyimpanan. Sedangkan, ketika kita ingin menggunakan gambar untuk keperluan web atau untuk email, kita perlu menggunakan gambar dengan resolusi yang lebih rendah agar file tidak terlalu besar dan mudah dikirim atau diunggah.

**Jawaban Tugas**


Garis Vertikal

```python
# Membuat garis vertikal ditengah image
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for x in range (tinggi):
  image_3[x,int((lebar)/2)] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Garis Menyilang Diagonal

```python
# Membuat garis diagonal
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for y in range (tinggi):
  image_3[y,int(y*(lebar/tinggi))] = [255,255,255]
  image_3[y,lebar-1-int(y*(lebar/tinggi))] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
